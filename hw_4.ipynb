__2.__ Проверить на совместность и выяснить, сколько решений будет иметь система линейных уравнений:

   а) $\begin{cases}
3x_{1}-x_{2}+x_{3}=4, \\
2x_{1}-5x_{2}-3x_{3}=-17, \\
x_{1}+x_{2}-x_{3}=0;
\end{cases}$

<b>Решение</b>

$\begin{pmatrix}
\left.\begin{matrix}
3 & -1 & 1 \\
2 & -5 & -3 \\
1 & 1 & -1 
\end{matrix}\right|
\begin{matrix}
4 \\

0
\end{matrix}
\end{pmatrix}$ = $\begin{pmatrix}
\left.\begin{matrix}
4 & 0 & 0 \\
2 & -5 & -3 \\
1 & 1 & -1 
\end{matrix}\right|
\begin{matrix}

\end{matrix}
\end{pmatrix}$ = $\begin{pmatrix}
\left.\begin{matrix}

\end{matrix}\right|
\begin{matrix}

\end{matrix}
\end{pmatrix}$ 

$\begin{pmatrix}
\left.\begin{matrix}

-1 & -8 & 0 \\
4 & 0 & 0
\end{matrix}\right|
\begin{matrix}
0 \\
-17 \\
4
\end{matrix}
\end{pmatrix}$ = $\begin{pmatrix}
\left.\begin{matrix}
1 & -1 & -1\\
0 & -8 & -1\\
0 & 0 & 4
\end{matrix}\right|
\begin{matrix}
0 \\
-17 \\

\end{matrix}
\end{pmatrix}$ => rank(A) = rank($\tilde{A}$) = n = 3 => совместна, 1 решение.

```python
A = np.array([[3, -1, 1], [2, -5, -3], [1, 1, -1]])
b = np.array([4, -17, 0])
A_ext = np.column_stack([A, b])

rank_A = np.linalg.matrix_rank(A)
rank_A_ext = np.linalg.matrix_rank(A_ext)
n = len(A[0])


def print_results(A, A_ext, rank_A, rank_A_ext, n):
    print(f'A:\n{A}\n')
    print(f'A_ext:\n{A_ext}\n')

    print(f'rank(A) = {rank_A}\n')
    print(f'rank(A_ext) = {rank_A_ext}\n')
    print(f'Количество переменных: {n}\n')

    if rank_A == rank_A_ext == n:
        print(f'Система имеет единственное решение.')
    elif rank_A == rank_A_ext < n:
        print(f'Система имеет бесконечное кол-во решений.')
    elif rank_A < rank_A_ext:
        print(f'Система несовместна.')

print_results(A, A_ext, rank_A, rank_A_ext, n)
```

__3*.__ Решить систему линейных уравнений методом $LU$-разложения

$$\begin{cases}
2x_{1}+x_{2}+3x_{3}=1 \\
11x_{1}+7x_{2}+5x_{3}=-6 \\
9x_{1}+8x_{2}+4x_{3}=-5
\end{cases}$$

```python
A = np.array([[2, 1, 3], [11, 7, 5], [9, 8, 4]])
b = np.array([1, -6, -5])

lu, piv = scipy.linalg.lu_factor(A)
x = scipy.linalg.lu_solve((lu, piv), b)

print(f'A:\n{A}\n')
print(f'b:\n{b}\n')
print(f'x:\n{x}\n')
print(f'Ax:\n{A @ x}\n')
print(f'Ax - b:\n{A @ x - b}\n')
```

__4*.__ Решить систему линейных уравнений методом Холецкого

$$\begin{cases}
81x_{1}-45x_{2}+45x_{3}=531 \\
-45x_{1}+50x_{2}-15x_{3}=-460 \\
45x_{1}-15x_{2}+38x_{3}=193
\end{cases}$$

```python
A = np.array([[81, -45, 45], [-45, 50, -15], [45, -15, 38]])
b = np.array([531, -460, 193])

K = scipy.linalg.cho_factor(A)
x = scipy.linalg.cho_solve(K, b)

print(f'A:\n{A}\n')
print(f'b:\n{b}\n')
print(f'x:\n{x}\n')
print(f'Ax:\n{A @ x}\n')
print(f'Ax - b:\n{A @ x - b}\n')
```
